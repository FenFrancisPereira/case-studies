# Graph of Male and Female Buyers

```python
plt.figure(figsize=(13,13))
male_vs_female=sns.countplot(x='New Car',hue='Buyer Gender',data=carsdata)
plt.title('Whether the cars bought by Males and Females were new or not')
plt.xlabel('New Car',size=20)
plt.ylabel('Counts',size=20)
plt.bar_label(male_vs_female.containers[0])
plt.bar_label(male_vs_female.containers[1])
plt.show()
```

```python
## So from the above figure it is clear that 10000 people have participated in sales,out of which 
## 5052 are females and  4948 are males. I
```

# Top 5 cars based on Selling Price

```python
plt.figure(figsize=(15,15))
plt.bar(carsdata['Make'],carsdata['Sale Price'])
plt.title('Cars and the Sales Price')
plt.xlabel('Cars')
plt.ylabel('Sale Price')
plt.xticks(rotation=90)
plt.yticks()
plt.grid()
plt.show()
```

```python
## Based on the above graph, top 5 cars based on selling price are:
1) Ferrari
2)Holden
3)Hillman
4)Nissan
5)Honda
```

# The least 5 cars based on Resell Price

```python
plt.figure(figsize=(15,15))
plt.bar(carsdata['Make'],carsdata['Resell Price'])
plt.title('Cars and Resell Price')
plt.xlabel('Cars')
plt.ylabel('Resell Price')
plt.xticks(rotation=90)
plt.yticks()
plt.grid()
```

```python
## Based on resell price, the Least 5 cars are:
1)MG
2)Panoz
3)Tesla
4)Corbin
5)Smart
```
